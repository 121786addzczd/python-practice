# 記述統計量②

Pythonを使って記述統計量を算出

1 記述統計量の算出


| 名前 | 点数 |
|------|------|
| Aさん | 45   |
| Bさん | 40   |
| Cさん | 55   |
| Dさん | 60   |
| Eさん | 65   |
| Fさん | 35   |

1. 平均値を算出しましょう。
2. 6人の偏差の和を算出しましょう。
3. 平方和を算出しましょう。
4. 分散を算出しましょう。
5. 標準偏差を算出しましょう。

### 1. 平均値を算出

```python
student_scores = {
    "Aさん": 45,
    "Bさん": 40,
    "Cさん": 55,
    "Dさん": 60,
    "Eさん": 65,
    "Fさん": 35,
}


# 点数の合計と生徒の人数を計算
total_score = sum(student_scores.values())
number_of_students = len(student_scores)

# 平均点を計算
average_score = total_score / number_of_students

# 平均点を出力
print(f"平均点: {average_score:.2f}")
```

### 2. 6人の偏差の和を算出しましょう。

```python
student_scores = {
    "Aさん": 45,
    "Bさん": 40,
    "Cさん": 55,
    "Dさん": 60,
    "Eさん": 65,
    "Fさん": 35,
}

# 点数の平均値を求める
average_score = sum(student_scores.values()) / len(student_scores)

# 偏差の和を計算
deviation_sum = 0  # 初期値を0と設定
for score in student_scores.values():
    deviation = score - average_score  # 偏差を計算
    deviation_sum += deviation

# 偏差の和を出力
print(f"偏差の和: {deviation_sum:.2f}")
```

### 3. 平方和を算出

```python
student_scores = {
    "Aさん": 45,
    "Bさん": 40,
    "Cさん": 55,
    "Dさん": 60,
    "Eさん": 65,
    "Fさん": 35,
}

# 点数の平均値を求める
average_score = sum(student_scores.values()) / len(student_scores)

# 平方和を計算
sum_of_squares = 0  # 初期値を0と設定
for score in student_scores.values():
    deviation = score - average_score  # 偏差を計算
    sum_of_squares += deviation ** 2  # 偏差の平方を合計に加える

# 平方和を出力
print(f"平方和: {sum_of_squares:.2f}")
```

### 4. 分散を算出

```python
student_scores = {
    "Aさん": 45,
    "Bさん": 40,
    "Cさん": 55,
    "Dさん": 60,
    "Eさん": 65,
    "Fさん": 35,
}

NUM_STUDENTS = len(student_scores)

# 点数の平均値を求める
average_score = sum(student_scores.values()) / NUM_STUDENTS

# 平方和を計算
sum_of_squares = 0  # 初期値を0と設定
for score in student_scores.values():
    deviation = score - average_score  # 偏差を計算
    sum_of_squares += deviation ** 2  # 偏差の平方を合計に加える

# 分散を計算
variance = sum_of_squares / NUM_STUDENTS
variance = int(variance * 10) / 10.0

# 分散を出力
print(f"分散: {variance:.1f}")
```

### 5. 標準偏差を算出

```python

import math

student_scores = {
    "Aさん": 45,
    "Bさん": 40,
    "Cさん": 55,
    "Dさん": 60,
    "Eさん": 65,
    "Fさん": 35,
}

NUM_STUDENTS = len(student_scores)

# 分散を計算
variance = sum_of_squares / NUM_STUDENTS

# 標準偏差を計算
standard_deviation = math.sqrt(variance)

# 小数点第一位まで求め、かつ、小数点第二位を切り捨て
truncated_standard_deviation = int(standard_deviation * 10) / 10.0

# 標準偏差を出力
print(f"標準偏差: {truncated_standard_deviation:.1f}")
```

## 変動係数の算出


| 名前 | 点数 | 標準偏差 |
|------|------|------|
| A組 | 40   | 10   |
| B組 | 90   | 20   |

変動係数を算出し、A組とB組どちらのバラつきが大きいか確認しましょう。


```python
group_test_score_data = [
    {"name": "A組", "score": 40, "standard_deviation": 10},
    {"name": "B組", "score": 90, "standard_deviation": 20},
]

# 変動係数 = 標準偏差 / 平均値
max_coefficient_of_variation = -1  # 初期値を設定
max_coefficient_group_name = ""  # 初期値を設定

for group_data in group_test_score_data:
    name = group_data["name"]
    score = group_data["score"]
    std_dev = group_data["standard_deviation"]
    
    coefficient_of_variation = std_dev / score if score != 0 else 0
    
    print(f"{name}の変動係数: {coefficient_of_variation:.2f}")
    
    if coefficient_of_variation > max_coefficient_of_variation:
        max_coefficient_of_variation = coefficient_of_variation
        max_coefficient_group_name = name

print(f"{max_coefficient_group_name}のバラつきが大きい")
```

## 尖度と歪度

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes, load_linnerud, fetch_california_housing
```

```python
data1 = fetch_california_housing()
df1 = pd.DataFrame(data1.data, columns=data1.feature_names)
```

```python
plt.figure(figsize=(10, 5))
for i, col in enumerate(df1.columns):
    plt.subplot(3, 3, i + 1)
    sns.kdeplot(df1[col])
plt.tight_layout()
plt.show()
```
