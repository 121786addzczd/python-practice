|Aさん| 45 |

|Eさん| 65 |

|Iさん| 50 |
|Jさん| 55 |
|Kさん| 40 |


この表を使ってPythonを使って計算します（使用するライブラリは問いません）。
1. 最大値を算出しましょう。
2. 最小値を算出しましょう。
3. 平均値を算出しましょう。
4. 中央値を算出しましょう。
5. 最頻値を算出しましょう。
6. 第一四分位数を算出しましょう。
7. 第三四分位数を算出しましょう。
8. 四分位範囲を算出しましょう。

```python
# ■ 問2　Pythonを使って記述統計量の算出
import numpy as np
from scipy import stats

# データのリスト
data = [45, 40, 55, 60, 65, 95, 40, 30, 50, 55, 40]

# 最大値
max_val = np.max(data)
print("最大値:", max_val)

# 最小値
min_val = np.min(data)
print("最小値:", min_val)

# 平均値
mean_val = np.mean(data)
print("平均値:", mean_val)

# 中央値
median_val = np.median(data)
print("中央値:", median_val)

# 最頻値
mode_val = stats.mode(data)
print(f"最頻値: {mode_val.mode}, 出現回数: {mode_val.count}")

# 第一四分位数
q1 = np.percentile(data, 25)
print("第一四分位数:", q1)

# 第三四分位数
q3 = np.percentile(data, 75)
print("第三四分位数:", q3)

# 四分位範囲
iqr = q3 - q1
print("四分位範囲:", iqr)
```

この表を使ってPythonを使って計算します（使用するライブラリは問いません）。
1. 最大値を算出しましょう。
2. 最小値を算出しましょう。
3. 平均値を算出しましょう。
4. 中央値を算出しましょう。
5. 最頻値を算出しましょう。
6. 第一四分位数を算出しましょう。
7. 第三四分位数を算出しましょう。
8. 四分位範囲を算出しましょう。

```python
# ■ 問5　Pythonを使って記述統計量の算出
import numpy as np
from scipy import stats

data = [45, 40, 55, 60, 65, 95, 40, 30, 50, 55, 40, 10]

# 最大値
max_val = np.max(data)
print("最大値:", max_val)

# 最小値
min_val = np.min(data)
print("最小値:", min_val)

# 平均値
mean_val = np.mean(data)
print("平均値:", mean_val)

# 中央値
median_val = np.median(data)
print("中央値:", median_val)

# 最頻値
mode_val = stats.mode(data)
print(f"最頻値: {mode_val.mode}, 出現回数: {mode_val.count}")

# 第一四分位数
q1 = np.percentile(data, 25)
print("第一四分位数:", q1)

# 第三四分位数
q3 = np.percentile(data, 75)
print("第三四分位数:", q3)

# 四分位範囲
iqr = q3 - q1
print("四分位範囲:", iqr)
```
