## 演算の計算

```python
df
```

```python
# 統計値も計算を行う
df['lifeExp'] + 3
```

```python
# データフレームはベクトル演算に対応している
2 * df['lifeExp']
```

```python
# 通常のpythonコードではベクトル演算には対応していない
[1, 2, 3] + 3
```

```python

# 統計値を計算
print(f"平均値:{df['lifeExp'].mean()}")   # 平均値
print(f"中央値:{df['lifeExp'].median()}") # 中央値
print(f"最小値:{df['lifeExp'].min()}")    # 最小値
print(f"最大値:{df['lifeExp'].max()}")    # 最大値
print(f"標準偏差:{df['lifeExp'].std()}")  # 標準偏差
print(f"95パーセント値:{df['lifeExp'].quantile(0.95)}") # 95パーセント値
```
