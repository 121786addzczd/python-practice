# Pythonを使った円グラフの作成方法

```python
"""値とラベルに下のデータを使用して、円グラフを作成"""
import matplotlib.pyplot as plt
%matplotlib inline

value = [10, 15, 25, 35]
label = ['tomato', 'carrot', 'potato', 'onion']

plt.pie(x=value, labels=label)
plt.show()
```

```python
# 円グラフの中にパーセントを表示
plt.pie(x=value, labels=label, autopct='%1.1f%%')
plt.show()
```

```python
# 円グラフの中にパーセントを小数点以下第2位まで表示
plt.pie(x=value, labels=label, autopct='%.2f%%')
plt.show()
```

```python
"""
値とラベルに下記リストのデータを使用して、完成イメージの通りに円グラフを作成。
・各値の全体に占める割合（％）を小数点以下切り捨てで表示。
・凡例を追加。
"""
value = [10, 15, 25, 35]
label = ['tomato', 'carrot', 'potato', 'onion']

plt.pie(x=value, labels=label, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
# 12時の位置から円グラフ表示
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
# 9時の位置から円グラフ表示
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=180)
plt.show()
```

```python
# 12時の位置から時計周りに円グラフ表示
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
# 円グラフの色を変更
colorlist = ["pink", "yellow", "skyblue", "orange"] # valueの値と同じ色の数にすること
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False, colors=colorlist)
plt.show()
```

```python
"""
値とラベルに下記リストのデータを使用して、完成イメージの通りに円グラフを作成。

・カラーマップの種類は、plt.get_cmap関数で'Pastel1'を取得。
・取得したカラーマップの最初から4番目までの色をnp.arange関数で指定。
"""
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

value = [10, 15, 25, 35]
label = ['tomato', 'carrot', 'potato', 'onion']

cmap = plt.get_cmap('Pastel1')
colors= cmap(np.arange(4)) # 最初から４番目の色を指定
plt.pie(x=value, labels=label, colors=colors)
plt.show()
```

```python
# カラー変更（カラーコード）
colorlist = ["#BAD7DB", "#B0C6B5", "#E0E1C3", "#F3D6D6"] 
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False, colors=colorlist)
plt.show()
```

```python
# 一部を切りとる carrot
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90,
        counterclock=False, colors=colorlist, explode=[0, 0.3, 0, 0])
plt.show()
```

```python
# 一部を切り、中心から離す carrot
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90,
        counterclock=False, colors=colorlist, explode=[0, 1, 0, 0])
plt.show()
```

```python
# 立体感を付ける
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90,
        counterclock=False, colors=colorlist, explode=[0, 0.3, 0, 0], shadow=True)
plt.show()
```

```python
# グラフにタイトルを付ける
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90,
        counterclock=False, colors=colorlist, explode=[0, 0.3, 0, 0], shadow=True)
plt.title('PieGraph')
plt.show()
```

```python
# グラフ保存（savefig）
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90,
        counterclock=False, colors=colorlist, explode=[0, 0.3, 0, 0], shadow=True)
plt.title('PieGraph')
plt.savefig('sample1.png')
```

```python
# ドーナツグラフ
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.pie([100], colors='white', radius=0.5)
plt.show()
```

```python
# %の位置調整（円の中心が０、円周が１）
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False, pctdistance=0.75)
plt.pie([100], colors='white', radius=0.5)
plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の売上金額に関するデータのcsvファイルを読み込んだもの。
df_sum1は商品分類ごとの売上金額を降順で並び替えたもの。
df_sum2は商品分類と商品名ごとに売上金額を集計したもの。
df_sum3は商品分類と商品名ごとに売上金額を集計し、商品分類ごとの売上金額の小計を追加したもの。
まず、外側に商品分類ごとの売上金額の円グラフを作成。
次に、内側に商品名ごとの売上金額の円グラフを作成。
最後に、最内に白い円グラフを作成。
二重ドーナツ型の円グラフを作成。

・外側の円グラフの色は、matplotlibのカラーマップの'tab10'の0, 1, 2番目の色とする。
・内側の円グラフの色は、matplotlibのカラーマップの'tab20c'の0, 1, 4, 5, 6, 9, 10番目の色とする。
・それぞれの円グラフは12時の位置から時計回りに作成する。
・内側の円グラフの半径を0.7とする。
・内側の円グラフのラベルの位置を中心から0.8とする。
・中心の白い円グラフの半径を0.5とする。
"""

import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
%matplotlib inline

# 1つ目のデータフレーム（外側の円グラフ）
df_sum1 = pd.read_csv('csv_files/Matplotlib-Seaborn07.csv')[['商品分類','売上金額']].groupby('商品分類').sum().sort_values(by='売上金額', ascending=False)
# 2つ目のデータフレーム
df_sum2 = pd.read_csv('csv_files/Matplotlib-Seaborn07.csv')[['商品分類', '商品名', '売上金額']].groupby(['商品分類', '商品名']).sum()
# 3つ目のデータフレーム（内側の円グラフ）
df_sum3 = pd.merge(df_sum2, df_sum1, left_index=True, right_index=True, suffixes=['', '小計']).sort_values(by=['売上金額小計', '売上金額'], ascending=False)

value1 = df_sum1['売上金額']
label1 = df_sum1.index
value2 = df_sum3['売上金額']
label2 = df_sum3.index.get_level_values('商品名')

# 色を設定
cmap1 = plt.get_cmap("tab10")
cmap2 = plt.get_cmap("tab20c")
color1 = cmap1([0, 1, 2])
color2 = cmap2([0, 1, 4, 5, 6, 9, 10])

# 外側の円グラフ
plt.pie(x=value1, labels=label1, startangle=90, 
        counterclock=False, colors=color1)
# 内側の円グラフ
plt.pie(x=value2, labels=label2, startangle=90, 
        counterclock=False, colors=color2, 
        radius=0.8, labeldistance=0.7)
# 中心の白い円の円グラフ
plt.pie([100], colors='white', radius=0.5)
plt.show()

```

### ↑の補足
pandasのmerge関数を使用するとき、データフレーム間で列名が重複している場合、それらを区別するために自動的に接尾辞が追加されます。しかし、suffixesパラメータを使用することで、重複した列名にどのような接尾辞を付けるかを自分で指定することができます。

具体的には、suffixes=('接尾辞1', '接尾辞2')の形で2つの接尾辞を指定します。ここで'接尾辞1'は左側のデータフレーム（merge関数の第一引数）の重複列に追加され、'接尾辞2'は右側のデータフレーム（merge関数の第二引数）の重複列に追加されます。

あなたのコードで言えば、suffixes=['', '小計']は、df_sum2（左側）とdf_sum1（右側）で重複する列名（ここでは'売上金額'）に、それぞれ''（空文字列）と'小計'を接尾辞として追加します。

その結果、df_sum2の'売上金額'列はそのまま'売上金額'という名前を保持し、df_sum1の'売上金額'列は'売上金額小計'という名前に変わります。これにより、結合後のデータフレームでは2つの異なる'売上金額'列を区別できます。
