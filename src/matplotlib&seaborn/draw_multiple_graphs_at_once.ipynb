# 複数グラフを同時に描画する方法

## ライブラリインポート

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## 二つのグラフをsubplotで表示

```python
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
# サインのグラフのみを表示
plt.plot(x, y_sin)
```

```python
# コサインのグラフのみを表示
plt.plot(x, y_cos)
```

```python
# 一つのグラフに二つのデータを表示
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
```

```python
# 二つのグラフを同時表示
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(121)
plt.plot(x, y_sin)
plt.subplot(122)
plt.plot(x, y_cos)
plt.show()
```

```python
# グラフ間の余白や位置を調整
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(121)
plt.plot(x, y_sin)
plt.subplot(122)
plt.plot(x, y_cos)

plt.subplots_adjust(wspace=0.5) # デフォルト値は0.2 グラフサイズは常に6:4
plt.show()
```

```python
# 全体のグラフサイズ変更
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(9,3))

plt.subplot(121)
plt.plot(x, y_sin)
plt.subplot(122)
plt.plot(x, y_cos)

plt.show()
```

```python
# それぞれのグラフにタイトルをつける
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(9,3))

plt.subplot(121)
plt.plot(x, y_sin)
plt.title('sin(x)')

plt.subplot(122)
plt.plot(x, y_cos)
plt.title('cos(y)')

plt.show()
```

```python
# グラフ全体にタイトルをつける
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(9,3))

plt.subplot(121)
plt.plot(x, y_sin)
plt.title('sin(x)')

plt.subplot(122)
plt.plot(x, y_cos)
plt.title('cos(y)')

plt.suptitle('Graphs', fontsize=18)
plt.subplots_adjust(top=0.8) # デフォルトでは0.9
plt.show()
```

```python
# 凡例を使い、上下のグラフスペースなし
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(9,4))
plt.subplot(211)
plt.plot(x, y_sin, color='coral', label='sin')
plt.legend()

plt.subplot(212)
plt.plot(x, y_cos, color='darkcyan', label='cos')
plt.legend()

plt.suptitle('graphs', fontsize=18)
plt.subplots_adjust(top=0.85, hspace=0) 
plt.show()
```

## Gridspecグラフ表示比率の変更

```python
from matplotlib import gridspec
```

```python
# グラフを縦に並べる
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize=(9,6))
grid= gridspec.GridSpec(nrows=2, ncols=1, height_ratios=[2,1]) # height_ratiosは高さの比率

ax1 = fig.add_subplot(grid[0])
ax1.plot(x, y_sin)

ax2 = fig.add_subplot(grid[1])
ax2.plot(x, y_cos)

plt.show()
```

```python
# グラフを横に並べる
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize=(9,6))
grid = gridspec.GridSpec(nrows=1, ncols=2, width_ratios=[2,1])

ax1 = fig.add_subplot(grid[0])
ax1.plot(x, y_sin)

ax2 = fig.add_subplot(grid[1])
ax2.plot(x, y_cos)

plt.show()
```

## subplotsを使ってよりすっきり書く方法

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(9,6))
# fig, ax = plt.subplots(2,2, figsize=(9,6)) # nrows, ncolsは省略できる

ax[0, 0].plot(x, y_sin)
ax[0, 1].plot(x, y_cos)
ax[1, 0].plot(x, y_sin)
ax[1, 1].plot(x, y_cos)

plt.show()
```

```python
# 縦に並べて表示
fig, ax = plt.subplots(4,1, figsize=(9,6))

ax[0].plot(x, y_sin)
ax[1].plot(x, y_cos)
ax[2].plot(x, y_sin)
ax[3].plot(x, y_cos)

plt.show()
```

# 軸の共有

```python
fig, ax = plt.subplots(2,2, figsize=(9,6), sharex=True, sharey=True) # グラフの間の軸ラベルを削除

ax[0, 0].plot(x, y_sin)
ax[0, 1].plot(x, y_cos)
ax[1, 0].plot(x, y_sin)
ax[1, 1].plot(x, y_cos)

plt.show()
```

## グラフタイトル追加

```python
fig, ax = plt.subplots(2,2, figsize=(9,6), sharex=True, sharey=True)

ax[0, 0].plot(x, y_sin)
ax[0, 0].set(title='sin')

ax[0, 1].plot(x, y_cos)
ax[0, 1].set(title='cos')

ax[1, 0].plot(x, y_sin)
ax[1, 1].plot(x, y_cos)

plt.suptitle('sin & cos Graphs', fontsize=18)
plt.show()
```

```python
"""
2つの波状の線グラフを1枚のグラフの中に作成。

・NumPyライブラリのlinspace関数を用いて負の円周率から正の円周率までの201個の等間隔の配列を作成し、x軸のデータとする。
・1つ目のグラフのy軸のデータは、x軸のデータの正弦（サイン）とする。
・2つ目のグラフのy軸のデータは、x軸のデータの余弦（コサイン）とする。
"""
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-np.pi, np.pi, 201)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.show()
```

```python
"""
2つの線グラフを縦に並べて作成。

・matplotlibライブラリのgridspecモジュールのGridSpecクラスを使用する。
・グラフ全体のサイズを横9インチ、縦6インチとする。
・グラフの表示を2行1列にして、上下のグラフの高さの比率を2：1とする。
・NumPyライブラリのlinspase関数を用いて負の円周率から正の円周率までの201個の等間隔の配列を作成し、x軸のデータとする。
・上のグラフのy軸のデータは、x軸のデータの正弦（サイン）とする。
・下のグラフのy軸のデータは、x軸のデータの余弦（コサイン）とする。
"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
%matplotlib inline

x = np.linspace(-np.pi, np.pi, 201)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize=(9,6))
grid= gridspec.GridSpec(2,1, height_ratios=[2,1]) 

ax1 = fig.add_subplot(grid[0])
ax1.plot(x, y_sin)

ax2 = fig.add_subplot(grid[1])
ax2.plot(x, y_cos)

plt.show()
```

```python
"""
x軸とy軸のデータに下記リストを使用して、完成イメージの通りに、3つの種類のグラフを横方向に作成。

・グラフ描画領域全体のサイズを横9インチ、縦3インチとする。
・1つ目のグラフは棒グラフ、2つ目のグラフは線グラフ、3つ目のグラフは散布図。
"""

import matplotlib.pyplot as plt
%matplotlib inline

names = ['A', 'B', 'C']
value = [10, 20, 30]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names, value)

plt.subplot(132)
plt.plot(names, value)

plt.subplot(133)
plt.scatter(names, value)

plt.suptitle('Three Graphs')
plt.show()
```

```python
"""
x軸とy軸のデータに下記リストを使用して、完成イメージの通りに、3つの種類のグラフを作成。

・上側のグラフは棒グラフ、下側左のグラフに線グラフ、下側右のグラフに散布図を表示。
・グラフ描画領域全体のサイズを横9インチ、縦3インチで設定。
・グラフのタイトルの文字サイズは16とする。
"""

import matplotlib.pyplot as plt
%matplotlib inline

names = ['A', 'B', 'C']
value = [10, 20, 30]

plt.figure(figsize=(9, 3))

plt.subplot(211)
plt.bar(names, value)

plt.subplot(223)
plt.plot(names, value)

plt.subplot(224)
plt.scatter(names, value)

plt.suptitle('Three Graphs', fontsize=16)
plt.show()
```

```python
"""
4つの線グラフを作成。

・NumPyライブラリのlinspase関数を用いて負の円周率から正の円周率までの201個の等間隔の配列を作成し、x軸のデータとする。
・左側の2つのグラフのy軸のデータは、x軸のデータの正弦（サイン）とする。
・右側の2つのグラフのy軸のデータは、x軸のデータの余弦（コサイン）とする。
"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize=(9,6))


ax1 = fig.add_subplot(221)
ax1.plot(x, y_sin)

ax2 = fig.add_subplot(222)
ax2.plot(x, y_cos)

ax3 = fig.add_subplot(223)
ax3.plot(x, y_sin)

ax4 = fig.add_subplot(224)
ax4.plot(x, y_cos)

plt.show()
```

```python
"""
plt.subplots関数を使用してaxesオブジェクトとfigureオブジェクトを同時に生成し、4つの線グラフを作成。

・NumPyライブラリのlinspase関数を用いて負の円周率から正の円周率までの201個の等間隔の配列を作成し、x軸のデータとする。
・グラフ描画領域全体のサイズを横9インチ、縦6インチで設定する。
・左側のグラフのy座標は、x座標の値の正弦（サイン）とする。
・右側のグラフのy座標は、x座標の値の余弦（コサイン）とする。
"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(2, 2, figsize=(9,6))

ax[0, 0].plot(x, y_sin)
ax[0, 1].plot(x, y_cos)
ax[1, 0].plot(x, y_sin)
ax[1, 1].plot(x, y_cos)

plt.show()
```

```python
"""
plt.subplots関数を使用して4つのグラフを作成し、表示。

・NumPyライブラリのlinspase関数を用いて負の円周率から正の円周率までの201個の等間隔の配列を作成し、x軸のデータとする。
・グラフ全体のサイズを横9インチ、縦6インチで設定。
・グラフ間の軸ラベルは削除するように（軸ラベルを共有するように）設定。
・左側のグラフのy座標は、x座標の値の正弦（サイン）とする。
・右側のグラフのy座標は、x座標の値の余弦（コサイン）とする。
"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-np.pi, np.pi, 201)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, ax = plt.subplots(2,2, figsize=(9,6), sharex=True, sharey=True) # グラフの間の軸ラベルを削除

ax[0, 0].plot(x, y_sin)
ax[0, 1].plot(x, y_cos)
ax[1, 0].plot(x, y_sin)
ax[1, 1].plot(x, y_cos)

plt.show()
```

```python
"""
plt.subplots関数を使用して、4つのグラフを作成。
全体のタイトルを'sin & cos Graphs'、左側のグラフのタイトルを'sin'、右側のグラフのタイトルを'cos'と設定し、完成イメージの通りに表示。

・NumPyライブラリのlinspase関数を用いて負の円周率から正の円周率までの201個の等間隔の配列を作成し、x軸のデータとする。
・グラフ全体のサイズを横9インチ、縦6インチで設定。
・グラフ間の軸ラベルは削除するように（軸ラベルを共有するように）設定。
・左側のグラフのy座標は、x座標の値の正弦（サイン）とする。
・右側のグラフのy座標は、x座標の値の余弦（コサイン）とする。
・全体のタイトルの文字の大きさは18とする。
"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-np.pi, np.pi, 201)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, ax = plt.subplots(2,2, figsize=(9, 6), sharex=True, sharey=True)

ax[0, 0].plot(x, y_sin)
ax[0, 0].set(title='sin')

ax[0, 1].plot(x, y_cos)
ax[0, 1].set(title='cos')

ax[1, 0].plot(x, y_sin)
ax[1, 1].plot(x, y_cos)

plt.suptitle('sin & cos Graphs', fontsize=18)
plt.show()
```
