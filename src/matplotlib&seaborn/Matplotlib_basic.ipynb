# Matplotlib

Matplotlib とはPythonのグラフを描くためのライブラリ。<br>
Matplotlibは非常に様々なグラフを描くことができる。<br>
たとえば、線グラフ、棒グラフのほかに、円グラフ・ヒストグラム・箱ひげ図・3Dグラフまで描くことができる。

```python
import matplotlib.pyplot as plt
# %matplotlib inline」を記述をすることで、JupyterLabにグラフを表示させることができる
%matplotlib inline 
```

```python
# 簡単なグラフを書く
x = [1, 2, 3, 4, 5] # x座標とy座標のデータの要素数は、一致しないとエラーになる
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.show() # plt.show関数でグラフを表示
```

```python
#（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
plt.plot(x, y)
plt.show()
```

```python
"""
グラフにラベルを表示させて、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータ。
完成イメージ
"""

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
グラフにラベルを設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータ。
・ラベルの文字サイズは18。
・ラベルの位置は、x軸は右端、y軸は上端。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel('X', fontsize=18, loc='right')
plt.ylabel('Y',  fontsize=18, loc='top')
plt.show()
```

```python
"""
下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
ポイント

・各座標の値は、リスト型のデータ。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
"""
グラフにグリッドを追加して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに。
・グリッドの色を'red'、スタイルを'dotted'、太さを1とする。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(color='red', linestyle='dotted', linewidth=1)
plt.show()
```

```python
"""
y軸のみにグリッドを追加して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(axis='y')
plt.show()
```

```python
"""
折れ線グラフの色を赤に設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
グラフの色をグレースケールの0.8に設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに。
・グレースケールは、色の濃淡を0.0から1.0までの数値で指定できる。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, color='0.8')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
# グレースケールとは、白と黒、そしてその間におけるグレーの色の種類を数値で表したもの。
# 黒は0.0、白は1.0を示す。
```

```python
"""
グラフの線の太さを5に設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
ポイント

・各座標の値は、リスト型のデータに。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, linewidth=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
折れ線グラフの線の太さを10、透明度を0.5に設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, linewidth=10, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
折れ線グラフのスタイルを点線に設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, linestyle='dotted')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
折れ線グラフのスタイルを一点鎖線に設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータにします。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, linestyle='dashdot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
円形のマーカーを設定し、座標下記の座標を通る折れ線グラフを作成
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
三角形のマーカーを設定して、下記の座標を通る折れ線グラフを作成
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='^')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
四角形のマーカーを設定して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータ。
・マーカーの大きさは10とする。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x,y, marker='s', markersize=10, markeredgecolor="black",markerfacecolor="white")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
グラフに注釈を追加し、下記の座標を通る線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータに
・注釈テキストの文字サイズは15。
・注釈の矢印先端の座標は(3,9)、注釈テキスト1文字目の座標は(1,20)
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.annotate('annotation', xy=(3,9), xytext=(1,20), fontsize='15', color='red', arrowprops=dict(color='gray'))
plt.show()
```

```python
"""
グラフに注釈を追加し、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータ
・注釈テキストの文字サイズは15
・注釈の線の先端の座標は(4,16)、注釈テキスト1文字目の座標は(2,20)
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.annotate('important', xy=(4,16), xytext=(2,20), fontsize='15', color='red', arrowprops=dict(color='gray', arrowstyle='-'))
plt.show()
```

```python
"""
グラフにテキストを追加して、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータ
・テキストを追加するには、plt.text関数を使用
・テキストの文字サイズは15
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.text(2, 20, 'important', fontsize=15)
plt.show()
```

```python
"""
グラフにタイトルをつけて、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)

・各座標の値は、リスト型のデータ
・タイトル名はgraph title
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('graph title')
plt.show()
```

```python
"""
グラフにタイトルをつけて、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
ポイント

・各座標の値は、リスト型のデータ。
・グラフのタイトルの位置は「-0.3」に指定。
"""

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('graph title', y=-0.3)
plt.show()
```

```python
"""
タイトルをつけて、下記の座標を通る折れ線グラフを作成。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
ポイント

・Matplotlibで日本語を表示させるためには、日本語のフォントをインポートする必要がある。
・タイトル名は'タイトル'
"""
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('グラフタイトル')
plt.show()
```

```python
"""
複数の折れ線グラフ(1)〜(3)を表示

・x座標とy座標のデータをNumpy配列にすることで、配列に対して四則演算ができるようになる。
・折れ線グラフ(1)は下記の座標を通るものとする。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
・折れ線グラフ(2)、(3)は、折れ線グラフ(1)のy座標をそれぞれ1/2、1/3した座標を通るものとする。
"""

import numpy as np
 
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,24])
plt.plot(x, y)
plt.plot(x, y/2)
plt.plot(x, y/3)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
複数の折れ線グラフ(1)〜(3)を表示。

・x座標とy座標のデータをNumpy配列にすることで、配列に対して四則演算ができるようになる。
・折れ線グラフ(1)は下記の座標を通るものとする。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
・折れ線グラフ(2)、(3)は、グラフ(1)のy座標をそれぞれ1/2、1/3した座標を通るものとする。
・折れ線グラフ(1)の色 'skyblue'
・折れ線グラフ(2)の色 'pink'
・折れ線グラフ(3)の色 'gray'
・x軸のラベル 'X'
・y軸のラベル 'Y'
"""

x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,24])
plt.plot(x, y, color='skyblue')
plt.plot(x, y/2, color='pink')
plt.plot(x, y/3, color='gray')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""
凡例をつけて、ひとつのグラフに複数の折れ線グラフ(1)〜(3)を表示。

・x座標とy座標のデータをNumpy配列にすることで、配列に対して四則演算ができるようになる。
・折れ線グラフ(1)は下記の座標を通るものとする。
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
・折れ線グラフ(2)、(3)は、グラフ(1)のy座標をそれぞれ1/2、1/3した座標を通るものとする。
"""

x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,24])
plt.plot(x, y, color='blue', label='A')
plt.plot(x, y/2, color='orange', label='B')
plt.plot(x, y/3, color='green', label='C')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
"""
完成イメージの通りに、凡例の文字サイズとグラフの線の色を変更し、ひとつのグラフに複数の折れ線グラフ(1)〜(3)を表示

・折れ線グラフ(1)は下記の座標を通るものとする
（x座標,y座標）= (1,1)(2,4)(3,9)(4,16)(5,25)
・折れ線グラフ(2)、(3)は、グラフ(1)のy座標をそれぞれ1/2、1/3した座標を通るものとする。
・凡例の文字サイズは15とする。
"""

x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,24])
plt.plot(x, y, color='purple', label='A')
plt.plot(x, y/2, color='orange', label='B')
plt.plot(x, y/3, color='blue', label='C')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(fontsize=15)
plt.show()
```
