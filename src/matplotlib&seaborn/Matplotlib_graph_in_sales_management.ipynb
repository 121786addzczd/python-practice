# 売上管理表で棒グラフでグラフ化

```python
"""
アパレル会社の売上のデータを、月ごとの売上金額に集計。
まず、データフレームの売上日と売上金額を、それぞれ任意の変数に代入。
そして、横軸を売上日、縦軸を売上金額として、棒グラフを作成。

・グラフ描画領域全体のサイズを横9インチ、縦6インチに設定。
・グラフの棒の幅は20。
・グラフを表示させる際に、y軸の目盛りを指数表記から通常の整数表記に変更。
"""

import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
# ↑notebook内にグラフを表示するための記述

# 月ごとの合計データフレーム作成
df = pd.read_csv('csv_files/sample_matplot.csv', index_col='売上日', parse_dates=True )[['売上金額']].resample('M').sum()
# 月ごとの売上棒グラフ
volume = df['売上金額']
date = df.index

plt.figure(figsize=(9, 6))
plt.bar(date, volume, width=20)
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
"""
アパレル会社の社員ごとの売上金額に関するデータが含まれているcsvファイルがある。
まず、このファイルを読み込み、'氏名'と'商品分類'ごとの'売上金額'の合計を算出。
次に、このデータフレームのインデックスが'商品分類'、カラムが'氏名'となるよう整形。
そして、Dataframeクラスのplotメソッドを使用して、'商品分類'と'氏名'ごとの'売上金額'の合計の棒グラフを作成。

・グラフ描画領域全体のサイズを横9インチ、縦6インチに設定。
・グラフ描画領域全体の文字サイズは12。
・軸ラベルの文字の大きさは15、タイトルの文字の大きさは18。
・get_level_valuesメソッドを使用すると、マルチインデックスから特定のインデックスの値のみを抽出。
・グラフを表示させる際に、y軸の目盛りを指数表記から通常の整数表記に変更。
"""

import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
%matplotlib inline

df = pd.read_csv('csv_files/sample_matplot.csv')[['氏名', '商品分類', '売上金額']].groupby(['氏名','商品分類']).sum()
df = df.unstack(level=0)

df.plot(kind='bar', rot=0, figsize=(9, 6), fontsize=12) # rot=0はラベル横並びにするオプション
label = df.columns.get_level_values('氏名')
plt.legend(label)
plt.xlabel('商品分類', fontsize=15)
plt.ylabel('売上金額（円）', fontsize=15)
plt.title('商品別売上/氏名', fontsize=18)
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

## 商品分類積み上げ棒グラフ

```python
"""
アパレル会社の社員ごとの売上金額に関するデータが含まれているcsvファイルがある。
まず、このファイルを読み込み、'氏名'と'商品分類'ごとの'売上金額'の合計を算出。
そして、Dataframeクラスのplotメソッドを使用して、グラフの色の変更と凡例の設定を行い、'氏名'と'商品分類'ごとの'売上金額'の合計の積み上げ棒グラフを作成。

・グラフ描画領域全体のサイズを横9インチ、縦6インチに設定。
・x軸のラベルは水平方向に表示。
・グラフを表示させる際に、y軸の目盛りを指数表記から通常の整数表記に変更。
"""

import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
%matplotlib inline

df = pd.read_csv('csv_files/sample_matplot.csv')[['氏名', '商品分類', '売上金額']].groupby(['氏名','商品分類']).sum()
df = df.unstack()

df.plot(kind='bar', rot=0, figsize=(9, 6), stacked=True) # rot=0はラベル横並びにするオプション
plt.legend(['アウター','トップス','ボトムス'])
plt.xlabel('氏名')
plt.ylabel('売上金額（円）')
plt.title('氏名/商品別売上')
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
"""
アパレル会社の社員ごとの売上金額に関するデータが含まれているcsvファイルがある。
まず、このファイルを読み込み、'氏名'と'商品分類'ごとの'売上金額'の合計を算出。
次に、このデータフレームのインデックスが'氏名'、カラムが'商品分類'となるように整形。
そして、Dataframeクラスのplotメソッドを使用して、凡例を設定し、'氏名'と'商品分類'ごとの'売上金額'の合計の積み上げ水平棒グラフを作成。

・グラフ描画領域全体のサイズを横9インチ、縦6インチに設定します。
・グラフを表示させる際に、x軸の目盛りを指数表記から通常の整数表記に変更します。
"""

df = pd.read_csv('csv_files/sample_matplot.csv')[['氏名', '商品分類', '売上金額']].groupby(['氏名','商品分類']).sum()
df = df.unstack()

df.plot(kind='barh', rot=0, figsize=(9, 6), stacked=True) # rot=0はラベル横並びにするオプション
plt.legend(['アウター','トップス','ボトムス'])
plt.xlabel('売上金額（円）')
plt.ylabel('氏名')
plt.title('氏名/商品別売上')
plt.ticklabel_format(style='plain',axis='x')
plt.show()
```
