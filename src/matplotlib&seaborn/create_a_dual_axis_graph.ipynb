```python
"""
アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、2軸グラフを作成。

・第1軸のグラフは棒グラフ、第2軸のグラフは折れ線グラフに設定。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

# データ読込
data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots()

ax1.bar(x, volume, color='darkcyan')

ax2 = ax1.twinx() # twinx x軸を共有してY軸を2軸に設定
ax2.plot(x, price, color='red')

plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'、第3軸のデータを'累計売上金額(円)'に設定して、3軸グラフを作成。

・第1軸は棒グラフ、第2軸と第3軸は折れ線グラフに設定。
・第3軸のy軸を、グラフの右側から1.2離れている距離に設定。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

# データ読込
data = pd.read_csv('csv_files/1month_data2.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']
total = data['累計売上金額(円)']

fig, ax1 = plt.subplots()

ax1.bar(x, volume, color='darkcyan')

ax2 = ax1.twinx() 
ax3 = ax1.twinx()

ax2.plot(x, price, color='red')
ax3.plot(x, total, color='orange')
ax3.spines['right'].set_position(('axes', 1.2)) # set_positionメソッドで、グラフの描画領域からの距離を1.2に設定

plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、2軸グラフを作成。

・グラフ領域のサイズを横9インチ、縦6インチで設定。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.bar(x, volume, color='darkcyan')

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')

plt.show()
```

```python
"""
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、2軸グラフを作成。

・グラフ領域のサイズを横9インチ、縦6インチで設定。
・グラフ領域の縦幅を1としたとき、グラフ領域の上端が0.6、グラフの下端が0.3に位置にする。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))
fig.subplots_adjust(top=0.6, bottom=0.3)

ax1.bar(x, volume, color='darkcyan')

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')

plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、タイトルとy軸ラベルを追加して2軸グラフを作成。

・タイトルの文字サイズは16、軸ラベルの文字サイズは14とする。
・グラフ領域のサイズを横9インチ、縦6インチで設定。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import japanize_matplotlib

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.bar(x, volume, color='darkcyan')
ax1.set_ylabel('数量（個）', fontsize=14)

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')
ax2.set_ylabel('売上金額（円）', fontsize=14)

plt.title('東京/売上金額', fontsize=16)
plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、タイトルと軸ラベルを追加して2軸グラフを作成。

・タイトルの文字サイズは16、軸ラベルの文字サイズは14とする。
・グラフ領域のサイズを横9インチ、縦6インチで設定。
・タイトルとx軸のラベルは、グラフとの間隔を10空けた位置に表示。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import japanize_matplotlib

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.set_xlabel('売上日',labelpad=10, fontsize=14) # グラフとの間隔を指定するには、引数labelpadを使用

ax1.bar(x, volume, color='darkcyan')
ax1.set_ylabel('数量（個）', fontsize=14)

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')
ax2.set_ylabel('売上金額（円）', fontsize=14)

plt.title('東京/売上金額',pad=10, fontsize=16)
plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、タイトルと軸ラベルを追加して2軸グラフを作成。

・タイトルの文字サイズは16、軸ラベルの文字サイズは14とする。
・タイトルとx軸のラベルは、グラフとの間隔を10空けた位置に表示。
・グラフ領域のサイズを横9インチ、縦6インチで設定。
・y軸のラベルは、グラフとの間隔を50開けた位置に、水平に表示。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import japanize_matplotlib

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.set_xlabel('売上日',labelpad=10, fontsize=14)

ax1.bar(x, volume, color='darkcyan')
ax1.set_ylabel('数量（個）', labelpad=50, rotation = "horizontal", fontsize=14)

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')
ax2.set_ylabel('売上金額（円）', labelpad=50, rotation = "horizontal", fontsize=14)

plt.title('東京/売上金額',pad=10, fontsize=16)
plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、2軸グラフを作成。

・タイトルの文字サイズは15、軸ラベルの文字サイズは14とする。
・グラフ領域のサイズを横9インチ、縦6インチで設定。
・x軸の目盛りの表示形式を「月/日」とします。(例：「1月1日」であれば、「01/01」と表示)

"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import japanize_matplotlib
import matplotlib.dates as dt

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.bar(x, volume, color='darkcyan')
ax1.set_ylabel('数量（個）', fontsize=14)

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')
ax2.set_ylabel('売上金額（円）', fontsize=14)

ax2.xaxis.set_major_formatter(dt.DateFormatter('%m/%d'))

plt.title('1月 東京/売上金額',pad=10, fontsize=16)
plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、2軸グラフを作成しましょう。

・タイトルの文字サイズは15、軸ラベルの文字サイズは14とする。
・グラフ領域のサイズを横9インチ、縦6インチで設定。
・x軸の目盛りの表示形式を「日」とします。（例：「1月1日」であれば、「01」と表示）

"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import japanize_matplotlib
import matplotlib.dates as dt

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.bar(x, volume, color='darkcyan')
ax1.set_ylabel('数量（個）', fontsize=14)

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')
ax2.set_ylabel('売上金額（円）', fontsize=14)

ax2.xaxis.set_major_formatter(dt.DateFormatter('%d'))

plt.title('1月 東京/売上金額',pad=10, fontsize=16)
plt.show()
```

```python
"""
下記のデータフレームは、アパレル会社の1ヶ月分の売上個数と売上金額に関するデータをまとめたcsvファイルを読み込み、'売上日'列を日付型データにして、インデックスに設定したもの。
完成イメージの通りに、x軸のデータを'売上日'、第1軸のデータを'数量(個)'、第2軸のデータを'売上金額(円)'に設定して、2軸グラフを作成。

・x軸の目盛りの表示形式を「日」とします。（例：「1月1日」であれば、「01」と表示）
・x軸に補助目盛りを表示。
・タイトルの文字サイズは15、軸ラベルの文字サイズは14とする。
・グラフ領域のサイズを横9インチ、縦6インチで設定。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import japanize_matplotlib
import matplotlib.dates as dt

data = pd.read_csv('csv_files/1month_data.csv', index_col='売上日', parse_dates=True)

x = data.index
volume = data['数量(個)']
price = data['売上金額(円)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.bar(x, volume, color='darkcyan')
ax1.set_ylabel('数量（個）', fontsize=14)

ax2 = ax1.twinx()
ax2.plot(x, price, color='red')
ax2.set_ylabel('売上金額（円）', fontsize=14)

ax2.xaxis.set_major_formatter(dt.DateFormatter('%d'))
ax2.set_xticks(x, minor=True)  # x軸に補助目盛りを表示

plt.title('1月 東京/売上金額',pad=10, fontsize=16)
plt.show()
```

```python
"""
下記のデータフレームは、ある日の東京の1日の気温、湿度、風速の時系列データをまとめたcsvファイルを読み込み、'Date'列を日付型データにして、インデックスに設定したもの。
x軸のデータを'Date'、第1軸のデータを'気温(℃)'、第2軸のデータを'湿度(%)'、第3軸のデータを'風速(m/s)'に設定して、3軸グラフを作成。

・第3軸のy軸を、グラフの右辺から1.1離れている位置に表示。
・グラフ領域のサイズを横9インチ、縦6インチで設定。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

kion_data = pd.read_csv('csv_files/kion_data2.csv', index_col='Date', parse_dates=True)

data = kion_data.index
temp = kion_data['気温(℃)']
humid = kion_data['湿度(%)']
wind = kion_data['風速(m/s)']

fig, ax1 = plt.subplots(figsize=(9,6))

ax1.plot(temp, color='coral')

ax2 = ax1.twinx()
ax3 = ax1.twinx()

ax2.plot(humid)
ax3.plot(wind, color='skyblue')
ax3.spines["right"].set_position(("axes", 1.1))  # 第3軸のy軸を、グラフの右辺から1.1離れている位置に表示

plt.show()
```
