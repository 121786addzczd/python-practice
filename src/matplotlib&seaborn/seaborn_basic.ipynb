# Seabornを使用した散布図、箱ひげ図、ヒストグラム等を描画する方法

Seabornでは、Matplotlibより美しいデザインのグラフを少ない記述で作成することができる

## ライブラリインポート

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## 使用データの準備

```python
sns.get_dataset_names()
```

```python
iris_data = sns.load_dataset('iris') # あやめのサンプルデータを使用
```

```python
iris_data # 左から順にがく片の長さ がく片の幅 花びらの長さ 花びらの幅 種の種類
```

## 散布図

```python
sns.relplot(x='sepal_length', y='petal_length', data=iris_data)
plt.show()
```

```python
# 比較のためにmatplotlobで同じ散布図を作成
plt.figure(figsize=(6,6))
plt.scatter(iris_data['sepal_length'], iris_data['petal_length'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
# seabornの方が簡単に少ない記述でかける
```

```python
# 背景色とグリッドを追加
sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data)
plt.show()
```

```python
# 白の背景に戻す
sns.set(style='white')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data)
plt.show()
```

```python
# 白の背景にグリッド付ける
sns.set(style='whitegrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data)
plt.show()
```

```python
# 点の色を好きな色に変更
sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, color='green')
plt.show()
```

```python
# カテゴリ別に色を分ける
sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species')  # speciesを指定
plt.show()
```

```python
# マーカのスタイルを変更
sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species', style='species') 
plt.show()
```

```python
# カテゴリ別にグラフを分けて表示
sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species', col='species') 
plt.show()
```

```python
# カテゴリごとの変量の散布図
sns.set(style='darkgrid')
sns.catplot(x='species', y='petal_length', data=iris_data) 
plt.show()
```

```python
# paletteで全体的な色見の変更
sns.set(style='darkgrid')
sns.catplot(x='species', y='petal_length', data=iris_data, palette='rocket') 
plt.show()
```

```python
# カラーパレットの指定
sns.set(style='darkgrid')
sns.catplot(x='species', y='petal_length', data=iris_data, palette='mako') 
plt.show()
```

```python
# カラーパレットの指定
sns.set(style='darkgrid')
sns.catplot(x='species', y='petal_length', data=iris_data, palette='pastel') 
plt.show()
```

```python
# カラーパレットの指定
sns.set(style='darkgrid')
sns.catplot(x='species', y='petal_length', data=iris_data, palette='dark') 
plt.show()
```

## 箱ひげ図


```python
sns.set(style='darkgrid')
sns.catplot(x='species', y='sepal_length', data=iris_data, kind='box')
plt.show()
```

## 線形回帰直線

```python
# がく片の長さと花びらの長さの相関の回帰直線
sns.set(style='darkgrid')
sns.regplot(x='sepal_length', y='petal_length', data=iris_data)
plt.show()
```

## 散布図とヒストグラム

```python
# 散布図の外側にヒストグラム作成(jointplot)
sns.set(style='darkgrid')
sns.jointplot(x='sepal_length', y='petal_length', data=iris_data)
plt.show()
```

```python
# 回帰直線と確率密度関数(reg)
sns.set(style='darkgrid')
sns.jointplot(x='sepal_length', y='petal_length', data=iris_data, kind='reg')
plt.show()
```

```python
# kindにhexと指定で六角形を作成
sns.set(style='white')
sns.jointplot(x='sepal_length', y='petal_length', data=iris_data, kind='hex')
plt.show()
```

```python
# カテゴリ別に色分け
sns.set(style='white')
sns.jointplot(x='sepal_length', y='petal_length', data=iris_data, hue='species')
plt.show()
```

## 複数の変量間の相関

```python
sns.set(style='darkgrid')
sns.pairplot(iris_data) # 全ての変量どうしの相関図
plt.show()
```

```python
# カテゴリ別に色分け
sns.set(style='darkgrid')
sns.pairplot(iris_data, hue='species')
plt.show()
```

```python
# ヒストグラムにする（diag_kindにhistを指定）
sns.set(style='darkgrid')
sns.pairplot(iris_data, hue='species', diag_kind='hist')
plt.show()
```

```python
# 全体のカラーはpaletteで変更
sns.set(style='whitegrid')
sns.pairplot(iris_data, hue='species', diag_kind='hist', palette='GnBu')
plt.show()
```

### マーカーの種類

マーカーは、seabornの`relplot()`関数を使用してデータを可視化する際に、データ点の形状を指定するために使用されます。以下は、seabornで使用可能な一部のマーカーの種類

| マーカー | 記号 |
|---------|------|
| サークル   | o    |
| 三角形 (下向き) | v    |
| 三角形 (上向き) | ^    |
| 正方形    | s    |
| 菱形     | D    |
| プラス記号  | +    |
| クロス記号  | x    |

これらのマーカーを`marker`引数に指定することで、データ点の外観を変更することができる

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を作成。

・グラフの描画にはseabornライブラリを使用。
・グラフのスタイルを「背景グレー、白のグリッド線あり」に設定。
・下記のカラムでマーカーの色を分けるようにする。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

iris_data = sns.load_dataset('iris')

sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species') 
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を作成。

・グラフの描画にはseabornライブラリを使用。
・グラフのスタイルを「背景白、グリッド線なし」に設定。
・'species'のカラムでマーカーの色を分け、マーカーの色を濃い色に設定。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

iris_data = sns.load_dataset('iris')

sns.set(style='white', palette='dark') # palette='dark'で色見を濃くする
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species')  # speciesを指定
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を作成。

・グラフの描画にはseabornライブラリを使用。
・グラフのスタイルを「背景グレー、白のグリッド線あり」に設定。
・'species'のカラムで、マーカーを'versicolor', 'virginica', 'setosa'の順番で色分け。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

iris_data = sns.load_dataset('iris')

sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species',  hue_order=['virginica', 'versicolor', 'setosa'])
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を作成。

・グラフの描画にはseabornライブラリを使用。
・グラフのスタイルを「背景グレー、白のグリッド線あり」に設定。
・'species'のカラムでマーカーの色と種類を設定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

iris_data = sns.load_dataset('iris')

sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species', style='species') 
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を品種ごとに作成。

・グラフの描画にはseabornライブラリを使用。
・グラフのスタイルを「背景グレー、白のグリッド線あり」に設定。
・'species'のカラムでマーカーの色と種類を設定。

"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

iris_data = sns.load_dataset('iris')

sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, hue='species', col='species') 
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を品種ごとに作成。

・グラフの描画にはseabornライブラリを使用。
・グラフのスタイルを「背景グレー、白のグリッド線あり」に設定。
・下記のカラムを参照して、マーカーの色と種類を設定。
・'virginica', 'versicolor', 'setosa'の順番で、マーカーの色分けと種類分けを行う。
・上記の順番で、マーカーの種類を「左向き三角形、下向き三角形、丸」とする。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

iris_data = sns.load_dataset('iris')

sns.set(style='darkgrid')
sns.relplot(x='sepal_length', y='petal_length', data=iris_data, 
            hue='species', hue_order=['virginica', 'versicolor', 'setosa'],
            style='species', style_order=['virginica', 'versicolor', 'setosa'], 
            markers=['<', 'v', '8'])
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を作成

・グラフの描画にはseabornライブラリを使用
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・x軸にカテゴリ変数、y軸に量的変数の下記のカラムを指定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

iris_data = sns.load_dataset('iris')

sns.set(style='darkgrid')
sns.catplot(x='species', y='petal_length', data=iris_data) 
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの散布図を作成。

・グラフの描画にはseabornライブラリを使用
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・x軸にカテゴリ変数、y軸に量的変数の下記のカラムを指定
・マーカーの全体的な色味を、'rocket'に設定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set(style='darkgrid')
sns.catplot(x='species', y='petal_length', data=iris_data, palette='rocket') 
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの分布図（箱ひげ図）を作成

・グラフの描画にはseabornライブラリを使用
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・x軸にカテゴリ変数、y軸に量的変数の下記のカラムを指定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set(style='darkgrid')
sns.catplot(x='species', y='sepal_length', data=iris_data, kind='box')
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットirisを使用して、あやめのデータの分布図（箱ひげ図）を作成

・グラフの描画にはseabornライブラリを使用
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・x軸にカテゴリ変数、y軸に量的変数の下記のカラムを指定
・箱ひげ図の全体の色味を'mako'に指定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set(style='darkgrid')
sns.catplot(x='species', y='sepal_length', data=iris_data, kind='box', col='species', palette='mako')
plt.show()
```
