# 「Seaborn」で色々なグラフを作成

## ライブラリのインポート

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## データセット用意

```python
tips_data = sns.load_dataset('tips')
tips_data # あるレストランで支払われたチップと関連データ 単位はドル
```

## 散布図

```python
sns.set(style='darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
# カテゴリごとのplot分ける
plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data, hue='time')
plt.show()
```

```python
# カテゴリごとに複数グラフで分ける（ランチとディナーを別々に）
sns.set(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips_data, col='time', kind='scatter')
plt.show()
```

```python
# 複数グラフのサイズを変更する
sns.set(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips_data, col='time', kind='scatter', height=3, aspect=2) # デフォルトではheight=5, aspect=1
plt.show()
```

## 回帰直線

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.regplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
# 散布図にカテゴリ別の回帰直線を表示（lmplotメソッド）
sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data, hue='smoker') # 禁煙者か喫煙者かでわける
plt.show()
```

```python
# lmplpotではグラフのサイズは変更できない
plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data, hue='smoker') # 禁煙者か喫煙者かでわける
plt.show()
```

```python
# relplotと同様height aspectで調整
sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data, hue='smoker', height=6, aspect=1.5)
plt.show()
```

## ヒストグラム

```python
plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data)
plt.show()
```

```python
# 性別で色分け
plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex')
plt.show()
```

```python
# 性別で色分けしたが見づらいため、積み上げグラフにする
plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='stack')
plt.show()
```

```python
# 重ねず表示
plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge')
plt.show()
```

```python
# 幅を設定 shrink
plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8)
plt.show()
```

```python
# グラフの色を変更
plt.figure(figsize=(9, 6))

sns.set_palette('Set1')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8)
plt.show()
```

```python
# グラフの色を変更（set_paletteメソッド使わない）
plt.figure(figsize=(9, 6))

sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8, palette='Set1')
plt.show()
```

```python
# カラーパレットを使用せずに独自のパレットを使う
plt.figure(figsize=(9, 6))

colors = ['coral', 'darkcyan']

sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8, palette=colors)
plt.show()
```

## 複数グラフ

```python
sns.set(style='darkgrid')
sns.pairplot(tips_data, hue='time', palette='Set1', diag_kind='hist')
plt.show()
```

```python
# height, aspectでサイズ変更
sns.set(style='darkgrid')
sns.pairplot(tips_data, hue='time', palette='Set1', diag_kind='hist', height=4, aspect=1)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、チップ金額の散布図を分けて作成

・グラフの描画にはseabornライブラリを使用
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・'time'列のLunchとDinnerでグラフを分けて表示
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.set(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips_data, col='time', kind='scatter')
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、チップ金額の散布図を作成

・グラフの描画にはseabornライブラリを使用
・各グラフのサイズは、高さを3インチ、縦に対する横の比率を2とする
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.set(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips_data, col='time', kind='scatter', height=3, aspect=2)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、チップ金額の散布図を作成

・グラフの描画にはseabornライブラリを使用
・各グラフのサイズは、高さを3、縦に対する横の比率を1とする
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.set(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips_data, row='time', kind='scatter', height=3, aspect=1) # rowで縦並びを指定
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、チップ金額の散布図を作成

・グラフの描画にはseabornライブラリを使用
・各グラフのサイズは、高さを3、縦に対する横の比率を1とする
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・LunchとDinnerは縦方向に、MaleとFemaleは横方向にグラフを分け
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.set(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips_data, row='time', col='sex', kind='scatter', height=3, aspect=1)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、チップ金額の回帰直線と散布図を作成

・グラフの描画にはseabornライブラリを使用
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.regplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、カテゴリ別のチップ金額の散布図と回帰直線を作成

・グラフの描画にはseabornライブラリを使用
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・'smoker'列のyesとnoで、マーカーと回帰直線の色を分け
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data, hue='smoker')
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、カテゴリ別のチップ金額の散布図と回帰直線を作成

・グラフの描画にはseabornライブラリを使用
・各グラフのサイズは、高さを6、縦に対する横の比率を1.5とする
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・'smoker'列のyesとnoで、マーカーと回帰直線の色を分ける
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data, hue='smoker', height=6, aspect=1.5)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、合計金額とチップ金額の散布図を作成
時間帯、性別でグラフを分け、完成イメージの通りに喫煙者と非喫煙者で色分けした回帰直線を作成

・グラフの描画にはseabornライブラリを使用します。
・グラフ描画領域全体のサイズを縦6インチ、アスペクト比1.5の比率で設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・時間帯(time)は縦方向に、性別(sex)は横方向にグラフを分ける
・'smoker'列のyesとnoで、マーカーと回帰直線の色を分ける
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data,
           hue='smoker', row='time', col='sex', height=6, aspect=1.5)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、曜日ごとのデータ数を表すヒストグラムを作成

・グラフの描画にはseabornライブラリを使用
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、曜日ごとのデータ数を表すヒストグラムを作成

・グラフの描画にはseabornライブラリを使用
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・MaleとFemaleで色分けを行う
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex')
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、曜日ごとのデータ数を表すヒストグラムを作成

・グラフの描画にはseabornライブラリを使用
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・MaleとFemaleで色分けを行い、積み上げ棒グラフにする
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='stack')
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、曜日ごとのデータ数を表すヒストグラムを作成

・グラフの描画にはseabornライブラリを使用
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・MaleとFemaleで色分けを行い、横並びにグラフを表示
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、曜日ごとのデータ数の累積を表す、累積ヒストグラムを作成

・グラフの描画にはseabornライブラリを使用します。
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・MaleとFemaleで色分けを行い、横並びにグラフを表示
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9, 6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8, cumulative=True) # 累積ヒストグラムにするには、引数cumulativeにTrueを渡す
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、曜日ごとのデータ数を表すヒストグラムを作成

・グラフの描画にはseabornライブラリを使用
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・カラーパレットに’Set1’を指定して、MaleとFemaleで色分けを行う
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9, 6))

sns.set_palette('Set1')
sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、曜日ごとのデータ数を表すヒストグラムを作成

・グラフの描画にはseabornライブラリを使用
・グラフ描画領域全体のサイズを縦6インチ、横9インチで設定
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・Maleの色を’coral'、Felameの色を'darkcyan'としてカラーパレットに指定し、色分けを行う
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')

plt.figure(figsize=(9, 6))

colors = ['coral', 'darkcyan']

sns.histplot(x='day', data=tips_data, hue='sex', multiple='dodge', shrink=0.8, palette=colors)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、すべての変量同士の相関グラフを表示

・グラフの描画にはseabornライブラリを使用
・各グラフのサイズは、高さを4、縦に対する横の比率を1とする
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・カラーパレットに’Set1’を指定して、LunchとDinnerで色分けを行う
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')
sns.set(style='darkgrid')
sns.set_palette('Set1')
sns.pairplot(tips_data, hue='time', palette='Set1', diag_kind='hist',  height=4, aspect=1)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用して、チップデータすべての変量同士の相関グラフを、重複の無いように表示

・グラフの描画にはseabornライブラリを使用
・各グラフのサイズは、高さを4、縦に対する横の比率を1とする
・グラフのスタイルを、「背景グレー、白のグリッド線あり」に設定
・カラーパレットに’Set1’を指定して、LunchとDinnerで色分けを行う
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')
sns.set(style='darkgrid')
sns.set_palette('Set1')
# corner=Trueと設定することで、pairplotの右上の三角部分（重複部分）を非表示にし、重複をなくしてpairplotが表示
sns.pairplot(tips_data, hue='time', diag_kind='hist',  height=4, aspect=1, corner=True)
plt.show()
```

```python
"""
seabornライブラリのサンプルデータセットtipsを使用
完成イメージの通りに、数値で表された変量同士の相関グラフを表示

・グラフの描画にはseabornライブラリを使用
・各グラフのサイズは高さを4、縦に対する横の比率を1とする
・グラフのスタイルを「背景グレー、白のグリッド線あり」に設定
・カラーパレットに’Set1’を指定
・'time'のカラムのLunchとDinnerで色分けを行う
・相関グラフに回帰直線を引く
・対角線上に'time'の値で分けた各変量のデータのヒストグラムを表示
"""

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips_data = sns.load_dataset('tips')
sns.set(style='darkgrid')
sns.set_palette('Set1')
# 相関グラフに回帰直線を引くには、引数kindに'reg'を渡す。
sns.pairplot(tips_data, hue='time', diag_kind='hist',  height=4, aspect=1, kind='reg', corner=True)
plt.show()
```
