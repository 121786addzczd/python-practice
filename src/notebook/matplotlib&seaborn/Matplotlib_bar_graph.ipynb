```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
下記2つのリストをX軸とY軸のデータに指定して。棒グラフを作成
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

plt.bar(l, v)
plt.show()
```

```python
"""
y軸のデータには、下記のリストを指定。
x軸のデータには、NumPyライブラリのarange関数で生成した「0からy軸のデータの要素数までで間隔が1の配列」を指定。
この2つのデータを使用して、完成イメージの通りに棒グラフを作成。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

v = [25, 32, 34, 20, 25, 15]
x = np.arange(len(v))
plt.bar(x, v)
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、グラフの太さを0.5に設定し、完成イメージの通りに棒グラフを作成
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

plt.bar(l, v, width=0.5)
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、完成イメージの通りに水平棒グラフを作成。
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

plt.barh(l, v)
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、グラフの色をカラーコードの'#A9C3AD'に設定し、完成イメージの通りに棒グラフを作成
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

plt.bar(l, v, color='#A9C3AD')
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、グラフの色をそれぞれ設定し、完成イメージの通りに棒グラフを作成。

・グラフA、C、D、Eの色を'silver'に設定する。
・グラフBの色を'coral'に設定する。
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]
color = ['silver', 'coral', 'silver', 'silver', 'silver'] 
plt.bar(l, v, color=color)
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、グラフの色をカラーコードの''#A9C3AD'、
グラフの枠線の色を'blue'に設定し、完成イメージの通りに棒グラフを作成
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]
plt.bar(l, v, color='#A9C3AD',  edgecolor='blue')
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、グラフの色をカラーコードの''#A9C3AD'、
グラフの枠線の色を'blue'、枠線の太さを3に設定し、完成イメージの通りに棒グラフを作成する。
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]
plt.bar(l, v, color='#A9C3AD',  edgecolor='blue', linewidth=3)
plt.show()
```

```python
"""
下記2つのリストをx軸とy軸のデータに指定して、それぞれの軸にラベルを追加し、完成イメージの通りに棒グラフを作成。
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]
plt.bar(l, v)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、それぞれの軸にラベルを追加し、完成イメージの通りに棒グラフを作成。

・軸ラベルの文字のプロパティは辞書型で指定。
　フォント'family'は'serif'、文字の色'color'は'saddlebrown'、文字の大きさ'size'は20とする。
 """

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]
plt.bar(l, v)

d = {'family': 'serf',
     'color': 'saddlebrown',
     'size': 20}

plt.xlabel('Category', fontdict=d )
plt.ylabel('Volume', fontdict=d)
# plt.xlabel('Category', size=20, color='saddlebrown', fontname='saddlebrown') # debug用
# plt.ylabel('Volume', size=20, color='saddlebrown', fontname='saddlebrown')  # debug用
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、グラフにタイトルを追加し、完成イメージの通りに棒グラフを作成
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

plt.bar(l, v)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.title('BarGraph')
plt.show()
```

```python
"""
下記2つのリストをx軸とy軸のデータに指定して、グラフタイトルを追加して、完成イメージの通りに棒グラフを作成。

・グラフタイトルの文字の設定は辞書型で指定。
　フォント'family'は'serif'、文字の色'color'は'b'、文字の大きさ'size'は20とする。
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

d = {'family': 'serf',
     'color': 'b',
     'size': 20}

plt.bar(l, v)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.title('BarGraph', fontdict=d)
plt.show()
```

```python
"""
2つのリストをx軸とy軸のデータに指定して、グラフ描画領域全体のサイズを横10インチ、縦6インチに設定し、完成イメージの通りに棒グラフを作成
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

plt.figure(figsize=(10, 6))
plt.bar(l, v)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.title('BarGraph')
plt.show()
```

```python
"""
下記2つのリストをx軸とy軸のデータに指定して、グラフ描画領域の設定を行い、完成イメージの通りに棒グラフを作成。
ポイント

・グラフ描画領域全体のサイズを横5インチ、縦5インチに設定。
・グラフの描画領域の背景色を'lightgray'に設定。
"""

l = ['A','B', 'C', 'D', 'E']
v = [10, 20, 30, 20, 10]

plt.figure(figsize=(5, 5), facecolor='lightgray')
plt.bar(l, v)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.title('BarGraph')
plt.show()
```

## 積み上げグラフ

```python
"""
3つのリストのデータから積み上げ棒グラフを作成。

・グラフの棒の幅は0.5とする。
"""
labels = ['A', 'B', 'C', 'D', 'E']
man = [35, 45, 25, 15, 20]
woman = [20, 25, 30, 40, 30]
width = 0.5

plt.bar(x, man, width, label='Man')
plt.bar(x, woman, width, bottom=man, label='Woman')
plt.xlabel('Category')
plt.ylabel('Number of People')
plt.title('Man and Woman')
plt.legend()
plt.show()
```

```python
"""
4つのリストのデータから積み上げ棒グラフを作成。

・グラフの棒の幅は0.5とする。
・appleとbananaのリストのデータはndarray型に変更し、要素同士を足し算する。
"""

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

l = ['Apr','May','Jun','Jul','Aug']
apple = [90,80,70,60,75]
banana = [50,40,40,65,70]
grape = [20, 15, 10, 5, 10]
width = 0.5

"""
grapeの開始位置には、appleとbananaのデータを合計した位置を指定。
そのためには、appleとbananaのリストのデータを、ndarray型にする必要がある。
"""
position = np.array(apple) + np.array(banana)

plt.bar(x, apple, width, label='apple')
plt.bar(x, banana, width, bottom=apple, label='banana')
plt.bar(x, grape, width, bottom=position, label='grape')
plt.xlabel('Month')
plt.ylabel('Number of Fruits')
plt.title('Fruits')
plt.legend()
plt.show()
```

## 水平積み上げ棒グラフ

```python
# 3つのリストのデータから水平積み上げ棒グラフを作成し、表示する

l = ['A', 'B', 'C', 'D', 'E']
man = [35,45,25,15,20]
woman = [20,25,30,40,30]
plt.barh(l, man, label='Man')
plt.barh(l, woman, label='Woman', left=man)
plt.xlabel('Category')
plt.ylabel('Number of People')
plt.title('Man and Woman')
plt.legend()
plt.show()
```

```python
"""
4つのリストのデータから100%水平積み上げ棒グラフを作成。

・apple,banana,grapeの各リストの値を、各月の3つの合計値に対する割合に変換する。
"""

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

l = ['Apr','May','Jun','Jul','Aug']
apple = [90,80,70,60,75]
banana = [50,40,40,65,70]
grape = [20, 15, 10, 5, 10]

s = np.array([apple, banana, grape]) 
s = np.divide(s, s.sum(axis=0)) * 100 # 配列の各要素の割合を計算

plt.barh(l, s[0], label='Apple')
plt.barh(l, s[1], label='Banana', left=s[0])
plt.barh(l, s[2], label='Grape', left=s[:2].sum(axis=0))

plt.xlabel('Rate')
plt.ylabel('Month')
plt.title('Fruits')
plt.legend(loc='upper left')
plt.show()
```

```python
# 3つのリストのデータから水平積み上げ棒グラフを作成(グラフの棒の高さは0.4とする)

l = ['A', 'B', 'C', 'D', 'E']
man = [35,45,25,15,20]
woman = [20,25,30,40,30]
height = 0.4

plt.barh(l, man, label='Man', height=0.4,)
plt.barh(l, woman, label='Woman', height=0.4, left=man)
plt.xlabel('Number of People')
plt.ylabel('Category')
plt.title('Man and Woman')
plt.legend()
plt.show()
```
